# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.



##Solução questão 1

Integrando a função e resolvendo:

\
$$\int_{0}^{2} x^5 (2 - x)dx$$ \

Precisamos expandir o integrando $x^5 (2 - x) = 2x^5 - x^6$ e resolver:

\
$$ \int_{0}^{2} (2x^5 - x^6)dx$$ \

Integrando cada termo:
- $$\int_{0}^{2} 2x^5  dx  = 2 \left( \frac{x^6}{6} \right) = \frac{2x^6}{6} = \frac{x^6}{3}$$
- $$\int_{0}^{2} x^6  dx = \frac{x^7}{7}$$

Somando as parciais e resolvendo:
\
$$ \left[ \frac{x^6}{3} \right]_{0}^{2} -  \left[ \frac{x^7}{7} \right]_{0}^{2}  $$ \
$$\left(\frac{2^6}{3} - \frac{0^6}{3}\right) -  \left( \frac{2^7}{7} - \frac{0^7}{7} \right)  $$ \
$$\frac{64}{3} - \frac{128}{7} =  $$ \
$$\frac{64}{21} ≈ 3.047 $$ \

Analisando resultado encontrado por código e gráfico:



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate

def F(x):
    return (2*x**5 - x**6)

# Definir o intervalo
x = np.linspace(0, 2, 400)

# Calcular os valores de F(x)
y = F(x)

# Plotar o gráfico da função
plt.plot(x, y, label=r'$F(x) = 2x^5 - x^6$', color='hotpink', linestyle='--')

# Hachurar a área sob a curva
plt.fill_between(x, y, where=[(0 <= xi <= 2) for xi in x], color='lavenderblush', alpha=0.4)

plt.title('Gráfico da função $F(x) = x^5(2 - x)$')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.legend()

plt.grid(True)
plt.show()

# Calcular a área sob a curva usando sympy
x_sym = symbols('x')
F_sym = x_sym**5 * (2 - x_sym)
area = integrate(F_sym, (x_sym, 0, 2))
print(f"A área sob a curva de F(x) no intervalo de 0 a 2 é: {area.evalf()}")
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

##Solução questão 3

Cosiderando o cálculo da integral:

\
$$v(t) = \int a(t)dt$$ \

Substituindo $a(t) = \cos(\pi t)$:

\
$$v(t) = \int \cos(\pi t)dt$$ \

Iremos substituir $u = \pi t$, $du = \pi dt$ e $dt = \frac{du}{\pi}$:

\
$$v(t) = \int \cos(u) \frac{du}{\pi}$$ \
$$v(t) =  \frac{1}{\pi}\int \cos(u) $$ \
$$v(t) =  \frac{1}{\pi}\ \sin(u) + C $$ \

Calculando $v(0)$:

\
$$v(t) =  \frac{1}{\pi}\ \sin(\pi t) + C $$ \
$$v(0) =  \frac{1}{\pi} \sin(0) + C $$ \
$$C =  \frac{1}{2\pi} $$ \

Então, $v(t)$:

\
$$v(t) =  \frac{1}{\pi}\ \sin(\pi t) +  \frac{1}{2\pi} $$ \

Para encontrar a distância, iremos calcular a integral entre os limites $[0, 1.5]$:

\

$$D =  \int_{0}^{1.5}\frac{1}{\pi}\ \sin(\pi t) +  \frac{1}{2\pi} dt$$ \
$$D =  \int_{0}^{1.5}\frac{1}{\pi}\ \sin(\pi t)dt +   \int_{0}^{1.5}\frac{1}{2\pi}dt $$ \
$$D = \frac{1}{\pi} \int_{0}^{1.5} -\cos(\pi t)dt + \int_{0}^{1.5}\frac{1t}{2\pi} $$ \
$$D = \left[ \frac{-\cos(\pi t)}{\pi^2} + \frac{1t}{2\pi} \right]_{0}^{1.5} $$ \
$$D = \left( \frac{-\cos(\pi 1.5)}{\pi^2} + \frac{1.5}{2\pi} \right) -  \left( \frac{-\cos(\pi 0)}{\pi^2} + \frac{0}{2\pi} \right)$$ \
$$D = \left(  \frac{1.5}{2\pi} \right) -  \left( \frac{1}{\pi^2}\right)$$ \
$$D ≈ 0.340 m $$ \

# Questão 4 - Vale até 1 ponto

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de usos de bibliotecas definidas para esta avaliação.  


##Solução questão 4

Nesse problema é necessário encontrar a integral da diferença entre as funções: $f(x) = -x^2 + 4x + 1$ e $g(x) = -x^3 + 7x^2 - 10x + 3$. Iremos calcular a integral

\
$$\int_{1}^{2} | g(x) - f(x)|dx$$ \

Simplificando a expressão da diferença de limies que será representada como $h$:

\
$$h(x) = (-x^3 + 7x^2 - 10x + 3) - (-x^2 + 4x + 1)$$ \
$$h(x) = (-x^3 + 7x^2 - 10x + 3) + x^2 - 4x - 1$$ \
$$h(x) = -x^3 + 8x^2 - 14x + 4$$ \

Iremos integrar a função $\int_{1}^{2} (-x^3 + 8x^2 - 14x + 4)dx \ $ abaixo, achando o valor $4.08333333333333$.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, lambdify

# Define as variáveis e as funções
x = symbols('x')
f = -x**2 + 4*x + 1
g = -x**3 + 7*x**2 - 10*x + 3

# Converte as funções sympy para funções que o numpy pode entender
f_lambdified = lambdify(x, f, "numpy")
g_lambdified = lambdify(x, g, "numpy")

# Define os limites de integração
x_lower = 1
x_upper = 2

# Define o número de retângulos para a soma de Riemann
num_rects = 100

# Define a largura de cada retângulo
width = (x_upper - x_lower) / num_rects

# Inicializa a soma total da área entre as curvas
total_area = 0

# Define os pontos intermediários no intervalo
x_vals = np.linspace(x_lower, x_upper, num_rects + 1)
midpoints = (x_vals[:-1] + x_vals[1:]) / 2

# Calcula a altura dos retângulos como a diferença entre f(midpoints) e g(midpoints)
heights = f_lambdified(midpoints) - g_lambdified(midpoints)

# Calcula a soma de Riemann
total_area = np.sum(heights * width)

# Para plotar a região delimitada
x_plot_vals = np.linspace(x_lower, x_upper, 400)
y_f_vals = f_lambdified(x_plot_vals)
y_g_vals = g_lambdified(x_plot_vals)

# Cria a figura e o eixo
fig, ax = plt.subplots()

# Plota as funções
ax.plot(x_plot_vals, y_f_vals, label=r"$f(x) = -x^2 + 4x + 1$", linestyle='--')
ax.plot(x_plot_vals, y_g_vals, label=r"$g(x) = -x^3 + 7x^2 - 10x + 3$", linestyle='--')

# Preenche a área entre as curvas
ax.fill_between(x_plot_vals, y_f_vals, y_g_vals, where=(y_f_vals > y_g_vals), color='skyblue', alpha=0.5)

# Desenha linhas verticais nos limites de integração
ax.axvline(x=x_lower, color='r', linestyle='--')
ax.axvline(x=x_upper, color='r', linestyle='--')
ax.legend()

# Desenha os retângulos
for i in range(num_rects):
    rect = plt.Rectangle((x_lower + i*width, g_lambdified(midpoints[i])), width, heights[i], color='skyblue', alpha=0.5)
    ax.add_patch(rect)

# Mostra o gráfico
plt.title(f"Área entre as curvas: {total_area:.5f}")
plt.grid()
plt.show()

# Calculando a integral da diferença com integrate
x = symbols('x')
f = -x**2 + 4*x + 1
g = -x**3 + 7*x**2 - 10*x + 3
h = f - g

area = abs(integrate(h, (x, 1, 2)))

print(f"A área limitada entre as curvas de 1 a 2 é: {area.evalf()}")
```

\
$$\int_{0}^{2} 16 - 6 + 5(e^2 - 1) = 10 + 5e^2 - 5 = 5 + 5e^2$$ \

\
$$M = \frac{1}{2-0} \int_0^2 (8x - 3 + 5e^{2-x}) \, dx = \frac{1}{2} \cdot 5(1 + e^2) = \frac{5}{2}(1 + e^2)$$ \
