##Solução questão 2:

a) $f(x)=x^4-4x^3+4x^2$

Para encontrar os pontos críticos da função $f(x)=x^4-4x^3+4x^2$, iremos calcular a derivada primeira, usando nos termos a regra da potência:

\
$$f'(x) = x^4-4x^3+4x^2$$ \
$$f'(x) = 4x^3 - 12x^2 + 8x$$ \

Agora, iremos igualar a derivada primeira a zero, porque os pontos críticos ocorrem onde a derivada de $f(x)$ é $0$:

\
$$4x^3 - 12x^2 + 8x = 0$$ \
$$4x(x^2 -3x + 2) = 0$$ \
$$4x(x - 1)(x - 2) = 0$$ \

Simplificando a função, descobrimos que os pontos críticos são $0, 1$ e $2$, pois se substituirmos $0$ em $4x$, $1$ em $x -1$ e $2$ em $x - 2$, chegaremos a $0$ em todas. Agora para descobrirmos quais são pontos de mínimo e quais são pontos de máximo, iremos calcular a derivada segunda, usando a regra da potência para os termos:

\
$$f''(x) = 4x^3 - 12x^2 + 8x$$ \
$$f''(x) = 12x^2 - 24x + 8$$ \

Substituindo os pontos $0, 1$ e $2$ na derivada segunda:

\
$$f''(0) = 12(0)^2 - 24(0) + 8 = 8 $$ \\
- $8 > 0$, logo é um ponto de minímo.

\
$$f''(1) = 12(1)^2 - 24(1) + 8 = -4$$ \\
- $-4 < 0$, logo é um ponto de máximo e um ponto de inflexão, onde a concavidade da função.


\
$$f''(2) = 12(2)^2 - 24(2) + 8 = 8$$ \
- $8 > 0$, logo é um ponto de minímo e um ponto de inflexão, onde a concavidade da função.


Então possuímos dois pontos de mínimo e um ponto de máximo. Agora iremos encontrar o $f(x)$ dos pontos  $0, 1$ e $2$:

\
$$f(0) = (0)^4-4(0)^3+4(0)^2 = 0$$ \

Para $ x = 0$, a coordenada é $(0, 0)$ para o primeiro ponto de mínimo.

\
$$f(1) = (1)^4-4(1)^3+4(1)^2 = 1$$ \

Para $ x = 1$, a coordenada é $(1, 1)$ para o ponto de máximo e um ponto de inflexão, onde a concavidade da função.

\
$$f(2) = (2)^4-4(2)^3+4(2)^2 = 0$$ \

Para $ x = 2$, a coordenada é $(2, 0)$ para o primeiro ponto de mínimo e um ponto de inflexão, onde a concavidade da função.

# Questão 3 - Vale até 4 Pontos

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

**Atenção**: O tempo é importante, então escolha, entre usar ferramentas computacionais ou gráficas. Independente da sua escolha você deverá escrever a resposta em uma célula de textos.


## Solução questão 3

a) $f(x)=e^{2x}−e^x$

O $-ln(2)$ é menor que zero, portanto é um ponto de máximo. O valor aproximado para $-log(2)$ é $-0.69423559$. Substituindo na primeira função para encontrarmos o $f(x)$, obteremos a coordenada $(-0.69423559, -0.2499997)$, como observamos no gráfico:







```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from math import e

x = sp.symbols('x')
f = sp.exp(2*x) - sp.exp(x)
f1 = sp.exp(2*x) - sp.exp(x)

# Calcula a primeira derivada de f(x) em relação a x usando sympy.diff()
derivate = sp.diff(f, x)
# Encontra os pontos críticos da função, onde a derivada é zero, usando sympy.solve()
critical_points = sp.solve(derivate, x)

# Calcula a segunda derivada de f(x) em relação a x usando sympy.diff()
second_derivate = sp.diff(derivate, x)
# Calcula os valores da segunda derivada nos pontos críticos
points = [second_derivate.subs(x, point) for point in critical_points]

# Itera sobre os pontos críticos e determina se são pontos de máximo, mínimo ou inflexão
for i, value in enumerate(points):
    if value > 0:
        print(f"Ponto de mínimo: {critical_points[i]}")
    elif value < 0:
        print(f"Ponto de máximo: {critical_points[i]}")
    else:
        print(f"Ponto de inflexão ou não está claro: {critical_points[i]}")

# Define uma função Python para f(x) usando numpy.exp() para a exponencial
def f(x):
    return np.exp(2*x) - np.exp(x)

# Gera valores de x uniformemente espaçados entre -3 e 5 usando numpy.linspace()
x_vals = np.linspace(-3, 5, 400)

# Calcula a derivada numérica de f(x) usando numpy.gradient()
derivada = np.gradient(f(x_vals), x_vals)
# Encontra os zeros da derivada numérica, indicando mudanças de sinal
pontos_zeros = x_vals[np.where(np.diff(np.sign(derivada)))[0]]

# Calculando primeiras e segundas derivadas
f1_primeiraDerivada = sp.diff(f1, x)
f1_segundaDerivada = sp.diff(f1_primeiraDerivada, x)

#Criando funções lambdified para uso no numpy e gráficos
f1_lambdified = sp.lambdify(x, f1, modules=['numpy'])
f1_primeiraDerivada_lambdified = sp.lambdify(x, f1_primeiraDerivada, modules=['numpy'])
f1_segundaDerivada_lambdified = sp.lambdify(x, f1_segundaDerivada, modules=['numpy'])

# Avaliando as funções no intervalo
f1_vals = f1_lambdified(x_vals)
f1_primeiraDerivada_vals = f1_primeiraDerivada_lambdified(x_vals)
f1_segundaDerivada_vals = f1_segundaDerivada_lambdified(x_vals)

# Plota o gráfico da função f(x) e marca os pontos críticos encontrados
plt.plot(x_vals, f(x_vals), label='$f(x)=e^{2x}−e^x$')
plt.title('Pontos críticos na função $f(x)=e^{2x}−e^x$')
plt.scatter(pontos_zeros, f(pontos_zeros), color='purple', label='Pontos Críticos')
plt.plot(x_vals, f1_primeiraDerivada_vals, label="$f'(x)$", linestyle='--')
plt.plot(x_vals, f1_segundaDerivada_vals, label="$f''(x)$", linestyle='-.')

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-3, 2)
plt.ylim(-0.5, 3)
plt.legend()
plt.grid(True)
plt.show()
```

##Solução questão 4

Para encontrar as dimensões do pôster com a menor área total, iremos definir o comprimento total como ($Ct$), largura do pôster ($Lp$) e altura do pôster ($Ap$)$$:

\
$$ Ct = Lp + Ap = (x+8) \cdot (x+12)$$

Sendo a área do material $x \cdot y = 384 cm^2$, a área total é:

\
$$ At = 384 + 12x + \frac{3072}{x} + 96$$ \

Agora, iremos derivar:

\
$$ \frac{d}{dx} (A)t = 12 - \frac{3072}{x^2} $$ \

Resolvendo a derivada primeira igualada a $0$, iremos encontrar $x = 16 cm$, resultado de uma raiz de $256$. Logo, encontramos $y = \frac{384}{x} = \frac{384}{16} = 24 cm$. Portanto, as dimensões do pôster com a menor área total são $16 cm$ de largura por $24 cm$ de altura.

**Rubrica de Avaliação**: até um ponto se as derivadas forem corretamente encontradas.

##Solução questão 5

Para resolver as derivadas parciais da função $ F(x, y) = x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y)$, primeiro iremos calcular a derivada de $F$ em relação a $x$ assumindo $y$ como uma constante, usando regra da soma de derivada, regra da potência, regra da derivação de constante e regra de derivação para funções trigonométricas e de cadeia em $\cos(x + 2y)$:

\
$$\frac{a}{ax} = x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y)$$ \
$$\frac{a}{ax} = \frac{a}{ax}(x^3) + \frac{a}{ax}(3x^2y) - \frac{a}{ax}(2xy^2) + y^3 + \frac{a}{ax}(\cos(x + 2y))$$ \
$$\frac{a}{ax} = 3x^2 + 6xy - 2y^2  -\sin(x + 2y)$$ \

Agora, iremos calcular $F$ em relação a $y$, assumindo $x$ como uma constante, usando regra da soma de derivada, regra da potência, regra da derivação de constante, regra do produto e regra de derivação para funções trigonométricas e de cadeia em $\cos(x + 2y)$:

\
$$\frac{a}{ay} = x^3 + 3x^2y - 2xy^2 + y^3 + \cos(x + 2y)$$ \
$$\frac{a}{ay} = \frac{a}{ay}(x^3) + \frac{a}{ay}(3x^2y) - \frac{a}{ay}(2xy^2) + \frac{a}{ay}(y^3) + \frac{a}{ay}(\cos(x + 2y))$$ \
$$\frac{a}{ay} = 3x^2 -4xy + 3y^2  -2\sin(x + 2y)$$ \
